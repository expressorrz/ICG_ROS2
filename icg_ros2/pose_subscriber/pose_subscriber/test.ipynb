```python
import numpy as np
import matplotlib.pyplot as plt
```

# Pick Aruco

```python

data_dict = np.load('data_pick_aruco.npz', allow_pickle=True)
poses_dict = data_dict['poses'].item()
twist_msg = data_dict['twists']

pose_keys = list(poses_dict.keys())

print(f'keys: {pose_keys}')
print(f'twist_msg shape: {twist_msg.shape}')
```

```python
print(f'poses_dict[{pose_keys[0]}] shape: {np.array(poses_dict[pose_keys[0]]).shape}')
poses = np.array(poses_dict[pose_keys[0]])
print(f'poses shape: {poses.shape}')
```

```python
plt.figure()
plt.plot(poses[:, 0], color='r', label='x')
plt.plot(poses[:, 1], color='g', label='y')
plt.plot(poses[:, 2], color='b', label='z')
plt.show()
```

```python
plt.figure()
plt.plot(twist_msg[:, 0], label='x')
plt.plot(twist_msg[:, 1], label='y')
plt.plot(twist_msg[:, 2], label='z')
plt.legend()
plt.title('Twist')
```
